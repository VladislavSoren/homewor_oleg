```python
def fun_sum_two_numbers(number1, number2):
    """
    number1, number2 - обязательные позиционные аргументы
    """
    result = number1 + number2
    return result
    # return number1 + number2
```

```python
fun_sum_two_numbers()
```

```python
fun_sum_two_numbers(1,2)
```

```python
def fun_sum_two_numbers(number1=0, number2=0):
    """
    number1, number2 - именованные аргументы (со значениями по умолчанию), т.е НЕ обяз-ые
    """
    return number1 + number2
```

```python
fun_sum_two_numbers()
```

```python
def inner_fun():
    print("inner_fun start")
    return "result inner_fun"

def fun(any_fun):
    print("fun start")
    result = any_fun()
    print("получили result из any_fun")
    return result
```

```python
fun(inner_fun)
```

```python
fun(inner_fun())  # inner_fun() -> "result inner_fun"
```
