```python
def get_rhomb(h, char = '-'):

    rhomb = ''

    for i in range(h):
        spaces = (h - i) - 1
        stars = (2 * i) + 1
        row = ' ' * spaces + char * stars + '\n'
        rhomb += row
    for i in range((h - 1) - 1, 0 - 1, -1):
        spaces = (h - i) - 1
        stars = (2 * i) + 1
        row = ' ' * spaces + char * stars + '\n'
        rhomb += row

    return rhomb
```

```python
result = get_rhomb(3)
print(result)
```

```python
def get_rhomb(h, char = '-', text = 'возвращаем ромб'):
    print(text)
    rhomb = ''

    for i in range(h):
        spaces = (h - i) - 1
        stars = (2 * i) + 1
        row = ' ' * spaces + char * stars + '\n'
        rhomb += row
    for i in range((h - 1) - 1, 0 - 1, -1):
        spaces = (h - i) - 1
        stars = (2 * i) + 1
        row = ' ' * spaces + char * stars + '\n'
        rhomb += row

    return rhomb
```

```python
result = get_rhomb(3)
print(result)
```
